```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("total_emission.csv")
df
```

```python
dfn = df
dfn.drop(columns=['Element', 'Item'], inplace=True)
```

```python
dfn['Element'] = df['Item'] + '_' + df['Element']
dfn
```

```python
unique_elm = dfn['Element'].unique()
unique_elm
```

```python
elm = [
    'Direct emissions (N2O)', 'Indirect emissions (N2O)',
       'Emissions (N2O)', 'Emissions (CO2eq) from N2O (AR5)',
       'Emissions (CO2eq) (AR5)', 'Emissions (CH4)',
       'Emissions (CO2eq) from CH4 (AR5)', 'Emissions (CO2)',
       'Emissions (CO2eq) from F-gases (AR5)'
    ]
```

```python
def find_net_amount(year, elm_no):
    year = str(year)
    x = np.sum(df[df['Element'] == elm_no][year])
    return x
```

```python
find_net_amount(2000, elm[1])
```

```python
results = []
for i in range(2000, 2021):
    result = {
        'year' : i,
        elm[0] : find_net_amount(i, elm[0]),
        elm[1] : find_net_amount(i, elm[1]),
        elm[2] : find_net_amount(i, elm[2]),
        elm[3] : find_net_amount(i, elm[3]),
        elm[4] : find_net_amount(i, elm[4]),
        elm[5] : find_net_amount(i, elm[5]),
        elm[6] : find_net_amount(i, elm[6]),
        elm[7] : find_net_amount(i, elm[7])
    }
    results.append(result)
results_df = pd.DataFrame(results)
results_df
```

```python
results_df.to_csv('co2_n2o_ch4.csv')
```
